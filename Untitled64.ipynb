```python
# Dates pour 2025
dates_2025 = pd.date_range('2025-01-01', pd.Timestamp.now(), freq='MS')

# Sites existants
sites = {'FANYE': fanye, 'NDIOL': ndiol}

bands_needed = ['B2','B3','B4','B8']

def add_indices_2025(image):
    image = image.select(bands_needed)
    ndvi = image.normalizedDifference(['B8','B4']).rename('NDVI')
    evi = image.expression(
        '2.5 * ((NIR - RED) / (NIR + 6 * RED - 7.5 * BLUE + 1))',
        {'NIR': image.select('B8'), 'RED': image.select('B4'), 'BLUE': image.select('B2')}
    ).rename('EVI')
    ndwi = image.normalizedDifference(['B3','B8']).rename('NDWI')
    return ee.Image.cat([ndvi, evi, ndwi])

rows_2025 = []

for site_name, geom in sites.items():
    for date in dates_2025:
        start = date.strftime('%Y-%m-%d')
        end = (date + pd.offsets.MonthEnd(1)).strftime('%Y-%m-%d')

        collection = (ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")
                      .filterBounds(geom)
                      .filterDate(start, end)
                      .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))
                      .map(add_indices_2025))

        if collection.size().getInfo() > 0:
            mean_image = collection.mean()
            values = mean_image.reduceRegion(
                reducer=ee.Reducer.mean(),
                geometry=geom,
                scale=10,
                maxPixels=1e9
            ).getInfo()

            row = {'Site': site_name, 'Date': start}
            row.update(values)
            rows_2025.append(row)

df_2025 = pd.DataFrame(rows_2025)
display(df_2025.head())
print(f'Données 2025 extraites : {len(df_2025)} enregistrements.')
```

```python
def add_ndvi(img):
    # Ensure the image has the required bands before computing NDVI
    ndvi = img.normalizedDifference(['B8', 'B4']).rename('NDVI')
    return ndvi.copyProperties(img, ['system:time_start'])

# Redefine the collection to ensure it contains the original spectral bands
# This assumes 'zone' is defined in a previous cell and refers to the intended geographic area.
collection = (ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
              .filterBounds(zone)
              .filterDate('2023-06-01', '2023-10-01') # Using the same date range as the original collection
              .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20)))

ndvi_collection = collection.map(add_ndvi)

# Moyenne par image
def extract_mean(img):
    stats = img.reduceRegion(
        reducer=ee.Reducer.mean(),
        geometry=zone,
        scale=10,
        maxPixels=1e9 # Augmenter le nombre maximal de pixels
    )
    return ee.Feature(None, {
        'date': img.date().format(),
        'NDVI': stats.get('NDVI')
    })

time_series = ndvi_collection.map(extract_mean)

# Convertir en liste Python
ts = time_series.getInfo()
```

```python
ee.Initialize(project='biomasse-senegal')

# ===================== 2. ZONES D’ETUDE =====================
sites = {
    'FANYE': ee.Geometry.Point([-15.88, 16.15]).buffer(5000),
    'NDIOL': ee.Geometry.Point([-16.25, 16.25]).buffer(5000)
}
```

```python
# ===================== 3. INDICES SPECTRAUX =====================
def add_indices(img):
    img = img.select(['B2','B3','B4','B8'])

    ndvi = img.normalizedDifference(['B8','B4']).rename('NDVI')

    evi = img.expression(
        '2.5*((NIR-RED)/(NIR+6*RED-7.5*BLUE+1))',
        {'NIR': img.select('B8'), 'RED': img.select('B4'), 'BLUE': img.select('B2')}
    ).rename('EVI')

    ndwi = img.normalizedDifference(['B3','B8']).rename('NDWI')

    return img.addBands([ndvi, evi, ndwi])
```

```python
# ===================== 4. EXTRACTION JOURNALIÈRE =====================
dates = pd.date_range('2025-10-01','2026-03-01',freq='D')

data = []

for site_name, roi in sites.items():
    for date in dates:

        start = date.strftime('%Y-%m-%d')
        end = (date + pd.Timedelta(days=1)).strftime('%Y-%m-%d') # End date is exclusive in Earth Engine

        collection = (ee.ImageCollection('COPERNICUS/S2_SR')
                      .filterBounds(roi)
                      .filterDate(start, end)
                      .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20)))

        # Map add_indices after filtering to avoid recomputing for each date if possible
        collection_with_indices = collection.map(add_indices)

        if collection_with_indices.size().getInfo() > 0:
            stats = (collection_with_indices.select(['NDVI','EVI','NDWI'])
                     .mean()
                     .reduceRegion(
                         reducer=ee.Reducer.mean(),
                         geometry=roi,
                         scale=10,
                         bestEffort=True)
                     .getInfo())

            if stats:
                stats['Date'] = date
                stats['Site'] = site_name
                data.append(stats)
```

```python
df_daily = pd.DataFrame(data)
df_daily['Date'] = pd.to_datetime(df_daily['Date'])
display(df_daily.head())
print(f'Daily data extracted: {len(df_daily)} records.')
```

```python
# ===================== 4. EXTRACTION MENSUELLE =====================
dates = pd.date_range('2025-10-01','2026-03-01',freq='MS')

data = []

for site_name, roi in sites.items():
    for date in dates:

        start = date.strftime('%Y-%m-%d')
        end = (date + pd.offsets.MonthEnd(1)).strftime('%Y-%m-%d')

        collection = (ee.ImageCollection('COPERNICUS/S2_SR')
                      .filterBounds(roi)
                      .filterDate(start, end)
                      .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20))
                      .map(add_indices))

        if collection.size().getInfo() > 0:
            stats = (collection.select(['NDVI','EVI','NDWI'])
                     .mean()
                     .reduceRegion(
                         reducer=ee.Reducer.mean(),
                         geometry=roi,
                         scale=10,
                         bestEffort=True)
                     .getInfo())

            stats['Date'] = date
            stats['Site'] = site_name
            data.append(stats)
```

### Préparation des Données pour le KNN
Nous allons préparer les données en définissant les caractéristiques (variables indépendantes) et la variable cible (variable dépendante). Pour notre analyse KNN, les indices spectraux (EVI, NDVI, NDWI) seront les caractéristiques, et le site (FANYE ou NDIOL) sera la variable cible.

```python
from sklearn.model_selection import train_test_split

# Définir les caractéristiques (features) et la variable cible (target)
X = df_daily[['EVI', 'NDVI', 'NDWI']]
y = df_daily['Site']

# Afficher les premières lignes des caractéristiques et de la cible pour vérification
print("Caractéristiques (X):")
display(X.head())
print("Variable Cible (y):")
display(y.head())
```

### Division des Données en Ensembles d'Entraînement et de Test
Pour évaluer la performance de notre modèle de manière fiable, nous divisons les données en un ensemble d'entraînement (utilisé pour entraîner le modèle) et un ensemble de test (utilisé pour évaluer le modèle sur des données non vues).

```python
# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Afficher les dimensions des ensembles résultants
print(f"Taille de l'ensemble d'entraînement (X_train): {X_train.shape}")
print(f"Taille de l'ensemble de test (X_test): {X_test.shape}")
print(f"Taille de la cible d'entraînement (y_train): {y_train.shape}")
print(f"Taille de la cible de test (y_test): {y_test.shape}")
```

### Entraînement du Modèle KNN
Maintenant que les données sont préparées, nous allons entraîner le modèle K-Nearest Neighbors (KNN) en utilisant les données d'entraînement. Nous choisirons un nombre de voisins (`n_neighbors`) pour commencer, par exemple 3.

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialiser le classifieur KNN
knn = KNeighborsClassifier(n_neighbors=3)

# Entraîner le modèle
knn.fit(X_train, y_train)

print("Modèle KNN entraîné avec succès!")
```

### Prédictions et Évaluation du Modèle KNN
Après l'entraînement, nous utiliserons le modèle pour faire des prédictions sur l'ensemble de test et évaluerons sa performance à l'aide de métriques telles que la précision (accuracy), le rapport de classification et la matrice de confusion.

```python
from sklearn.inspection import permutation_importance

# Calculer l'importance par permutation
result = permutation_importance(knn, X_test, y_test, n_repeats=10, random_state=42, n_jobs=-1)

# Organiser les résultats dans un DataFrame
feature_importances = pd.DataFrame({
    'Feature': X.columns,
    'Importance_Mean': result.importances_mean,
    'Importance_Std': result.importances_std
})

# Trier par importance moyenne décroissante
feature_importances = feature_importances.sort_values(by='Importance_Mean', ascending=False)

print("Importance des caractéristiques (Permutation Importance):")
display(feature_importances)
```

### Interprétation de l'Importance par Permutation

Ce tableau montre l'importance de chaque caractéristique pour le modèle KNN, calculée en mesurant la baisse de performance du modèle lorsque les valeurs de la caractéristique sont mélangées aléatoirement. Une valeur d'importance plus élevée indique que la caractéristique est plus cruciale pour les prédictions du modèle.

Dans notre cas, les caractéristiques qui ont la plus grande importance moyenne sont celles qui, lorsqu'elles sont "brouillées", ont le plus grand impact négatif sur la précision du modèle. Cela nous aide à comprendre quelles caractéristiques contribuent le plus à la distinction entre les sites FANYE et NDIOL.

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Faire des prédictions sur l'ensemble de test
y_pred = knn.predict(X_test)

# Évaluer la précision du modèle
accuracy = accuracy_score(y_test, y_pred)
print(f"Précision du modèle KNN : {accuracy:.2f}")

# Afficher le rapport de classification
print("\nRapport de Classification:")
print(classification_report(y_test, y_pred))

# Afficher la matrice de confusion
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=knn.classes_, yticklabels=knn.classes_)
plt.xlabel('Prédit')
plt.ylabel('Réel')
plt.title('Matrice de Confusion')
plt.show()
```

### Entraînement d'un Réseau de Neurones

Nous allons entraîner un réseau de neurones pour la classification des sites. Pour cela, nous utiliserons la bibliothèque Keras (intégrée à TensorFlow), qui est idéale pour construire et entraîner des modèles d'apprentissage profond. Étant donné la petite taille du jeu de données, un modèle simple sera utilisé.

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Standardiser les caractéristiques (important pour les réseaux de neurones)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# La variable cible doit être encodée numériquement pour les réseaux de neurones
# Nous utiliserons LabelEncoder comme pour XGBoost
le = LabelEncoder()
y_train_nn = le.fit_transform(y_train)
y_test_nn = le.transform(y_test)

# Définir le modèle de réseau de neurones
model = keras.Sequential([
    keras.layers.Input(shape=(X_train_scaled.shape[1],)), # Couche d'entrée avec le nombre de caractéristiques
    keras.layers.Dense(10, activation='relu'), # Une couche cachée avec 10 neurones et activation ReLU
    keras.layers.Dense(5, activation='relu'),  # Une autre couche cachée plus petite
    keras.layers.Dense(1, activation='sigmoid') # Couche de sortie pour la classification binaire (sigmoid)
])

# Compiler le modèle
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Afficher un résumé du modèle
model.summary()

print("Modèle de réseau de neurones défini et compilé avec succès!")
```

### Entraînement du Modèle de Réseau de Neurones

Nous allons maintenant entraîner le modèle en utilisant les données d'entraînement standardisées et les étiquettes encodées. Nous utiliserons un petit nombre d'époques en raison de la taille limitée du jeu de données.

```python
# Entraîner le modèle
history = model.fit(X_train_scaled, y_train_nn,
                    epochs=50,
                    batch_size=1, # Petite taille de lot en raison du petit dataset
                    validation_data=(X_test_scaled, y_test_nn),
                    verbose=0)

print("Entraînement du réseau de neurones terminé.")

# Visualiser la progression de l'entraînement
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy du Modèle')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Perte du Modèle')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

### Interprétation des Résultats de la Validation Croisée

Les scores de validation croisée fournissent une meilleure indication de la performance généralisée de votre modèle. Au lieu d'un seul score sur un ensemble de test spécifique, vous obtenez plusieurs scores (un pour chaque "pli"), ce qui aide à comprendre la stabilité du modèle.

*   **Précision moyenne:** Représente la performance moyenne du modèle sur les différents sous-ensembles de données.
*   **Écart-type:** Indique la variabilité des performances du modèle. Un écart-type faible suggère que le modèle est stable et que ses performances ne varient pas beaucoup en fonction des données d'entraînement/test.

Si la précision moyenne est élevée et l'écart-type faible, cela confirme que votre modèle est robuste et performant.
