```python
import os
import argparse
from typing import Optional, Dict, Tuple

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

# ─────────────────────────────────────────────────────────────────────────────
# 1. MODULES DE BASE (Encodeur/Décodeur U-Net)
# ─────────────────────────────────────────────────────────────────────────────

class ConvBlock(nn.Module):
    """Bloc standard : Convolution + Normalisation + Activation."""
    def __init__(self, in_ch, out_ch, kernel_size=3, norm=True, act=True):
        super().__init__()
        layers = [nn.Conv2d(in_ch, out_ch, kernel_size, padding=kernel_size // 2, bias=not norm)]
        if norm:
            layers.append(nn.BatchNorm2d(out_ch))
        if act:
            layers.append(nn.GELU())
        self.block = nn.Sequential(*layers)

    def forward(self, x):
        return self.block(x)

class DownBlock(nn.Module):
    """Bloc de l'encodeur : Convolution suivie d'une réduction de résolution (stride 2)."""
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = ConvBlock(in_ch, out_ch)
        self.down = nn.Conv2d(out_ch, out_ch, kernel_size=2, stride=2)

    def forward(self, x):
        skip = self.conv(x) # Connexion résiduelle pour le skip-connection
        return self.down(skip), skip

class UpBlock(nn.Module):
    """Bloc du décodeur : Sur-échantillonnage + Convolution avec concaténation du skip."""
    def __init__(self, in_ch, skip_ch, out_ch):
        super().__init__()
        self.up   = nn.ConvTranspose2d(in_ch, in_ch // 2, kernel_size=2, stride=2)
        self.conv = ConvBlock(in_ch // 2 + skip_ch, out_ch)

    def forward(self, x, skip):
        x = self.up(x)
        if x.shape != skip.shape:
            x = F.interpolate(x, size=skip.shape[2:], mode='bilinear', align_corners=False)
        x = torch.cat([x, skip], dim=1)
        return self.conv(x)

# ─────────────────────────────────────────────────────────────────────────────
# 2. MODULES DE FUSION ET MODÈLE FINAL
# ─────────────────────────────────────────────────────────────────────────────

class FiLMBlock(nn.Module):
    """Modulation linéaire des caractéristiques (FiLM) basée sur les données in-situ."""
    def __init__(self, c_dim: int, x_channels: int):
        super().__init__()
        self.gamma_proj = nn.Linear(c_dim, x_channels)
        self.beta_proj  = nn.Linear(c_dim, x_channels)

    def forward(self, x, c):
        # Calcul des paramètres de mise à l'échelle (gamma) et de décalage (beta)
        gamma = self.gamma_proj(c).unsqueeze(-1).unsqueeze(-1)
        beta  = self.beta_proj(c).unsqueeze(-1).unsqueeze(-1)
        return (1 + gamma) * x + beta

class BiomassLoss(nn.Module):
    """Fonction de perte combinant MSE spatiale et MAE sur les points terrain."""
    def __init__(self, w_mse=1.0, w_insitu=2.0):
        super().__init__()
        self.w_mse = w_mse
        self.w_insitu = w_insitu

    def forward(self, preds, target, target_mask, insitu_preds, insitu_target, insitu_mask):
        # Erreur quadratique moyenne sur la carte complète
        mse = F.mse_loss(preds * target_mask, target * target_mask)
        # Erreur absolue moyenne sur les points de mesure terrain
        mask_t = insitu_mask.unsqueeze(-1).float()
        mae_insitu = (F.l1_loss(insitu_preds * mask_t, insitu_target * mask_t, reduction='sum') / mask_t.sum().clamp(min=1))
        return {"loss": self.w_mse * mse + self.w_insitu * mae_insitu, "mse": mse, "mae_insitu": mae_insitu}

class BiomassDataset(Dataset):
    """Dataset synthétique pour simuler Sentinel-2 et données terrain."""
    def __init__(self, n_samples=100, H=64, W=64, context_length=6, target_length=4):
        self.n_samples = n_samples
        self.H, self.W = H, W
        self.context_length = context_length
        self.target_length = target_length

    def __len__(self):
        return self.n_samples

    def __getitem__(self, idx):
        # Génération de données aléatoires pour simuler le pipeline
        return {
            "satellite": torch.rand(self.context_length, 10, self.H, self.W),
            "static": torch.rand(3, self.H, self.W),
            "insitu_values": torch.rand(20, 3) * 50,
            "insitu_coords": torch.rand(20, 2),
            "insitu_mask": (torch.rand(20) > 0.2).float(),
            "target": torch.rand(self.target_length, 1, self.H, self.W) * 80,
            "target_mask": torch.ones(self.target_length, 1, self.H, self.W)
        }
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

@torch.no_grad()
def evaluate_detailed_metrics(model, loader, device):
    """Calcule des métriques détaillées (RMSE, MAE, R2) sur un ensemble de données."""
    model.eval()
    all_preds = []
    all_targets = []

    for batch in loader:
        batch = {k: v.to(device) for k, v in batch.items()}
        preds, _ = model(batch)

        # On aplatit pour calculer les scores sur tous les pixels valides (masque == 1)
        mask = batch['target_mask']
        p = preds[mask == 1].cpu().numpy()
        t = batch['target'][mask == 1].cpu().numpy()

        all_preds.append(p)
        all_targets.append(t)

    y_pred = np.concatenate(all_preds)
    y_true = np.concatenate(all_targets)

    # Calcul des scores statistiques classiques
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print(f"\n{'='*30}")
    print(f" MÉTRIQUES DE VALIDATION")
    print(f"{'='*30}")
    print(f"RMSE : {rmse:.4f} t/ha")
    print(f"MAE  : {mae:.4f} t/ha")
    print(f"R2   : {r2:.4f}")
    print(f"{'='*30}")

    return {"rmse": rmse, "mae": mae, "r2": r2}
```

### Baseline : Modèle Random Forest

Ici, nous transformons le problème spatial en un problème de régression tabulaire pixel par pixel pour comparer les performances de l'apprentissage profond avec une approche classique par Forêt Aléatoire.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

def prepare_tabular_data(dataset, n_samples=50):
    """Convertit les données d'image en format tabulaire (X, y) pour le RF."""
    X_list = []
    y_list = []

    for i in range(min(n_samples, len(dataset))):
        sample = dataset[i]
        # On concatène les bandes satellite (moyennées sur le temps) et les données statiques
        sat_mean = sample['satellite'].mean(0).reshape(10, -1) # [10, H*W]
        static = sample['static'].reshape(3, -1)             # [3, H*W]

        features = torch.cat([sat_mean, static], dim=0).T.numpy() # [H*W, 13]
        target = sample['target'][0, 0].reshape(-1).numpy()       # [H*W]

        X_list.append(features)
        y_list.append(target)

    return np.concatenate(X_list), np.concatenate(y_list)

# Initialisation d'un dataset pour le RF (puisque train_ds était local à main())
train_ds_rf = BiomassDataset(n_samples=50, H=64, W=64)

# Préparation des données
print("Préparation des données pour le Random Forest...")
X_rf, y_rf = prepare_tabular_data(train_ds_rf, n_samples=30)
X_train, X_test, y_train, y_test = train_test_split(X_rf, y_rf, test_size=0.2, random_state=42)

# Entraînement
rf_model = RandomForestRegressor(n_estimators=50, max_depth=10, n_jobs=-1, random_state=42)
rf_model.fit(X_train, y_train)

# Évaluation
y_pred = rf_model.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred)
r2_rf = r2_score(y_test, y_pred)

print(f"\n{'='*30}")
print(f" RÉSULTATS RANDOM FOREST")
print(f"{'='*30}")
print(f"RMSE : {np.sqrt(mse_rf):.4f} t/ha")
print(f"R2   : {r2_rf:.4f}")
print(f"{'='*30}")
```

```python
def plot_rf_comparison(model, dataset, sample_idx=0):
    """Reconstruit l'image à partir des prédictions tabulaires du RF pour comparaison."""
    sample = dataset[sample_idx]
    H, W = sample['target'].shape[2], sample['target'].shape[3]

    # Préparation des données tabulaires pour ce sample précis
    sat_mean = sample['satellite'].mean(0).reshape(10, -1) # [10, H*W]
    static = sample['static'].reshape(3, -1)             # [3, H*W]
    features = torch.cat([sat_mean, static], dim=0).T.numpy()

    # Prédiction
    rf_preds_flat = model.predict(features)
    rf_map = rf_preds_flat.reshape(H, W)
    target_map = sample['target'][0, 0].numpy()

    # Affichage
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    im1 = axes[0].imshow(target_map, cmap='viridis', vmin=0, vmax=80)
    axes[0].set_title('Vérité Terrain (Target)')
    fig.colorbar(im1, ax=axes[0])

    im2 = axes[1].imshow(rf_map, cmap='viridis', vmin=0, vmax=80)
    axes[1].set_title('Prédiction Random Forest')
    fig.colorbar(im2, ax=axes[1])

    plt.tight_layout()
    plt.show()

# Visualisation sur un échantillon de test
plot_rf_comparison(rf_model, train_ds_rf, sample_idx=5)
```
